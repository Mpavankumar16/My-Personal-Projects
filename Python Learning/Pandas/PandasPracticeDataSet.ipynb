## Concatenate 2 Different Series

```python
countryDataCount = df['Country'].value_counts()
countryDataCount
```

```python
finalResult = pd.concat([countryDataCount, peopleKnowPythonCount], axis='columns')
finalResult.rename(columns = {'count': 'TotalUsersCount', 'LanguageWorkedWith': 'PeopleKnowPythonCount'},inplace=True)
finalResult
```

```python
finalResult['%Count'] = (finalResult['PeopleKnowPythonCount'] / finalResult['TotalUsersCount'])*100
finalResult.sort_values(by='Country', inplace=True)
finalResult
```

## Cleaning Data

```python
import numpy as np
#If we don't have any info for a record then we will remove that record => Not having data means (None, numpy.nan)
df.dropna() # Will remove the records where some data is Na or Nan

# How=any will remove the record if any one key is having no value
# How=all will remove the record if all the key are no value
# index or columns means how to filter the data based on rows or Columns wise
df.dropna(axis='index', how='any') 

#subset is to check for only particular columns which we want to consider like => We want email for sure
df.dropna(axis='index', how='any', subset=['Hobbyist', 'OpenSource']) #In this case is any value is missing it will drop (OR) All -> AND

#Handling not req or none or other values
df.isna() # will give T or F DF with each cell whether it is a na or not => we can use replace to replace them
df.replace("NA", np.nan) #If we know the value then we replace it like this
df.fillna(0) #Fill all na values to be as 0

#DataTypes
df.dtypes #Will give us column type

df['ConvertedComp'] = df['ConvertedComp'].astype(float) #Their might be some values which look number but they are Str Need to convt them to only Float
```

```python
df['YearsCode'].head()
#df['YearsCode'].mean()
#df['YearsCode'] = df['YearsCode'].astype(float) #Their might be some values which have Less than 1 year as data
df['YearsCode'].unique()
```

```python
df['YearsCode'].replace('Less than 1 year', 0, inplace=True)
df['YearsCode'].replace('More than 50 years', 51, inplace=True)
df['YearsCode'].replace(np.nan, 0, inplace=True)
df['YearsCode'].unique()
```

```python
df['YearsCode'] = df['YearsCode'].astype(float)
df['YearsCode'].median()
```

```python
#Formats can be found here https://docs.python.org/3/library/datetime.html#strftime-and-strptime-behavior
#As our date are not formatted as req we are passing format
ethdf['Date'] = pd.to_datetime(ethdf['Date'], format='%Y-%m-%d %I-%p')
ethdf.head(2)
```
